```python
import matplotlib.pyplot as plt
import pandas as pd
import klib

# reading csv file with pandas
data = pd.read_csv('austin_weather.csv')
data = pd.DataFrame(data, columns=['Date', 'TempAvgF', 'TempHighF', 'TempLowF'])

# preparing data
x = list(data.tail(15)['Date'])
y = list(data.tail(15)['TempAvgF'])


#customizing plot size
plt.figure(figsize=(20, 9), dpi=150)

#add x and y + customizing MARKER size and ...
plt.plot(x,y, 'o--k', ms=15, mfc='r')

#add title and labels
plt.title('Climate change', {'color':'red', 'size': 50})
plt.xlabel("Dates", {'color':'blue', 'size': 30})
plt.ylabel("Average Temperature °F", {'color':'blue', 'size': 30})

plt.show()


```
